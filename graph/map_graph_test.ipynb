## NetworkX graph

```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

nodes = pd.read_csv('./graph-models/models/nodes_v0.2.csv')
edges = pd.read_csv('./graph-models/models/edges_v0.2.csv')

tuple_nodes = list(map(tuple, nodes.to_records(index=False)))
tuple_edges = list(map(tuple, edges.to_records(index=False)))

g = nx.Graph()
for node in tuple_nodes:
    g.add_node(node[2], pos=(node[0], node[1]))
for edge in tuple_edges:
    g.add_edge(edge[0], edge[1])
# g.add_nodes_from(tuple_nodes[:][2], pos=(tuple_nodes[:][0], tuple_nodes[:][1]))
# g.add_edges_from(tuple_edges)
nx.draw(g, with_labels=True, font_weight='bold')
```

```python
pe = edges.merge(nodes.rename(columns = {'x':'sourceX', 'y': 'sourceY', 'nodeId': 'source'}), on='source')
pe = pe.merge(nodes.rename(columns = {'x':'targetX', 'y': 'targetY', 'nodeId': 'target'}), on='target')

fig = plt.figure(figsize=(30,30))
plt.scatter(nodes['x'], nodes['y'], s=5, zorder=1)
plt.plot([pe['sourceX'], pe['targetX']], [pe['sourceY'], pe['targetY']], c='black',zorder=1)
for node in nodes.to_numpy():
    plt.text(node[1] + 20, node[2] + 20, node[0],size=6,c='red')
    plt.text(node[1], node[2] - 20, node[3] + "_" + node[4], size=7)
```
