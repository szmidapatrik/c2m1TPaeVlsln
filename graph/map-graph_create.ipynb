### Create graph

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("./_other/demo-parse/parse/player_frames/katowice2023_faze-vs-fnatic-m2-inferno.csv")
data2 = pd.read_csv("./_other/demo-parse/parse/player_frames/katowice2023_g2-vs-big-m2-inferno.csv")
data3 = pd.read_csv("./_other/demo-parse/parse/player_frames/katowice2023_g2-vs-faze-m1-inferno.csv")
data4 = pd.read_csv("./_other/demo-parse/parse/player_frames/katowice2023_g2-vs-liquid-m2-inferno.csv")
data5 = pd.read_csv("./_other/demo-parse/parse/player_frames/katowice2023_heroic-vs-natus-vincere-m1-inferno.csv")
data6 = pd.read_csv("./_other/demo-parse/parse/player_frames/katowice2023_heroic-vs-og-m3-inferno.csv")


pts1 = data[['x','y','z']].copy()
pts2 = data2[['x','y','z']].copy()
pts3 = data3[['x','y','z']].copy()
pts4 = data4[['x','y','z']].copy()
pts5 = data5[['x','y','z']].copy()
pts6 = data6[['x','y','z']].copy()

pts = pd.concat([pts1, pts2, pts3, pts4, pts5, pts6])
pts['numpts'] = 0

pts
```

```python
from scipy.spatial.distance import cdist
import numpy as np

x = np.linspace(-2000,3000, 60)
y = np.linspace(-1000,4000, 60)

grid = np.column_stack((np.repeat(x, len(y)), np.tile(y, len(x))))

dens = []

# Density
for x_axis in range(-2000,3000,1000):
    for y_axis in range(-1000,4000,1000):
        temp_dens = pts.loc[(pts['x'] >= x_axis) & (pts['x'] < x_axis + 1000) & (pts['y'] >= y_axis) & (pts['y'] < y_axis + 1000)].shape[0]
        dens.append([x_axis, y_axis, temp_dens])

for d in dens:
    pts.loc[(pts['x'] >= d[0]) & (pts['x'] < d[0] + 1000) & (pts['y'] >= d[1]) & (pts['y'] < d[1] + 1000), 'numpts'] = d[2]

pts['dens'] = pts['numpts'] / len(pts)

bars = pts[['numpts', 'dens']].drop_duplicates()
bars['numpts'] = bars['numpts'].astype(str)
plt.bar(bars['numpts'], bars['dens'])

pts_copy = pts.copy()

pts
```

```python
fig, ax = plt.subplots(1,2,figsize=(22,10))


# First plot
ax[0].scatter(pts['x'], pts['y'], s=1, c='darkslateblue', alpha=pts['dens'] / 3)
ax[0].scatter(grid[:,0], grid[:,1], s=0.1, c='black')
for y_l in range(-2000,4000,1000):
    ax[0].plot([y_l,y_l], [-1000, 4000], c="grey", linewidth=0.5)
for x_l in range(-1000,5000,1000):
    ax[0].plot([-2000, 3000], [x_l,x_l], c="grey", linewidth=0.5)
# Manually selected squares*
ax[0].fill_between([-2000, -1000], 0, 1000, color='lightblue', alpha=0.5)
ax[0].fill_between([-1000, 0], -1000, 0, color='lightblue', alpha=0.5)
ax[0].fill_between([0, 1000], -1000, 0, color='lightblue', alpha=0.5)
ax[0].fill_between([2000, 3000], 2000, 3000, color='lightblue', alpha=0.5)



ax[1].scatter(pts['x'], pts['y'], s=1, c='darkslateblue', alpha=1)
ax[1].scatter(grid[:,0], grid[:,1], s=0.1, c='black')
for y_l in range(-2000,4000,1000):
    ax[1].plot([y_l,y_l], [-1000, 4000], c="grey", linewidth=0.5)
for x_l in range(-1000,5000,1000):
    ax[1].plot([-2000, 3000], [x_l,x_l], c="grey", linewidth=0.5)
```

```python
x_sparse = np.linspace(-2000,3000, 30)
y_sparse = np.linspace(-1000,4000, 30)

grid_sparse = np.column_stack((np.repeat(x_sparse, len(y_sparse)), np.tile(y_sparse, len(x_sparse))))

# Function to calculate points inside circle
def count_points_in_circle(center, radius, points):
    distances = np.sqrt((points.iloc[:,0] - center[0]) ** 2 + (points.iloc[:,1] - center[1]) ** 2)
    points_inside_circle = np.sum(distances <= radius)
    return points_inside_circle

# Szűrés a grid pontokra
filtered_grid_points = []

for grid_point in grid_sparse:
    points_inside = count_points_in_circle(grid_point, 30, pts[['x', 'y']])
    if points_inside >= 5:
        filtered_grid_points.append(grid_point)

filtered_grid_points = np.array(filtered_grid_points)

graph_sparse = pd.DataFrame(filtered_grid_points, columns = ['gX', 'gY'])
#graph_sparse.to_csv('./graph-models/models/nodes_v0.2.csv', index=False)

fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(pts['x'], pts['y'], s=1, c='lightblue', alpha=1)
ax.scatter(graph_sparse['gX'], graph_sparse['gY'], s=0.1, c='black')
```

```python
# Spares grid
sprs1 = graph_sparse.loc[(graph_sparse['gX'] >= -2000) & (graph_sparse['gX'] < -1000) & (graph_sparse['gY'] >= 0) & (graph_sparse['gY'] < 1000)]
sprs2 = graph_sparse.loc[(graph_sparse['gX'] >= -1000) & (graph_sparse['gX'] <  1000) & (graph_sparse['gY'] >= -1000) & (graph_sparse['gY'] < 0)]
sprs3 = graph_sparse.loc[(graph_sparse['gX'] >=  2000) & (graph_sparse['gX'] <  3000) & (graph_sparse['gY'] >= 2000) & (graph_sparse['gY'] < 3000)]
sprs = pd.concat([sprs1, sprs2, sprs3])

# Dense grid
dense = pd.read_csv('./graph-models/test/test_0.4_2 (beta_v1).csv')
dense = dense.drop(dense.loc[(dense['gX'] >= -2000) & (dense['gX'] < -1000) & (dense['gY'] >= 0) & (dense['gY'] < 1000)].index)
dense = dense.drop(dense.loc[(dense['gX'] >= -1000) & (dense['gX'] <  1000) & (dense['gY'] >= -1000) & (dense['gY'] < 0)].index)
dense = dense.drop(dense.loc[(dense['gX'] >=  2000) & (dense['gX'] <  3000) & (dense['gY'] >= 2000) & (dense['gY'] < 3000)].index)

# Graph
graph_diffdens = pd.concat([sprs, dense])

fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(pts['x'], pts['y'], s=1, c='lightblue', alpha=1)
ax.scatter(graph_diffdens['gX'], graph_diffdens['gY'], s=0.1, c='black')
```
