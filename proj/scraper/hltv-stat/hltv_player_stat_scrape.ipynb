## HLTV Player Stat Scrape

Using the *HLTV_Player_Stat_Scraper*, download player stats for years.

```python
import sys
import os

sys.path.append(os.path.abspath('../../../package'))

from HLTV import HLTV_Player_Stat_Scraper
```

### 2023 Big Event stats

```python
FILE_TXT = 'url_done_2023.txt'
ERROR_TXT = 'url_error_2023.txt'
SAVE_CSV = '../../../data/player-stats/scraped-in-2024/2023/player_stats_2023.csv'

HLTV_Player_Stat_Scraper().get_player_stats(
    'https://www.hltv.org/stats/players?csVersion=CS2&startDate=2023-01-01&endDate=2023-12-31&matchType=BigEvents&gender=Men&minMapCount=0', 
    FILE_TXT,
    ERROR_TXT,
    SAVE_CSV, 
    2023
)
```
