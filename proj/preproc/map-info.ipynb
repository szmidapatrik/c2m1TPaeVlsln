```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import os
```

### Distribution of maps

```python
file_list_2023 = [file for file in os.listdir('../../data/matches-raw/cs2/2023') if file.endswith('.dem')]
file_list_2024 = [file for file in os.listdir('../../data/matches-raw/cs2/2024') if file.endswith('.dem')]

# Concat the two lists
file_list = file_list_2023 + file_list_2024

inferno_count = len([file for file in file_list if 'inferno' in file and 'p1' not in file and 'p2' not in file])
mirage_count = len([file for file in file_list if 'mirage' in file and 'p1' not in file and 'p2' not in file])
nuke_count = len([file for file in file_list if 'nuke' in file and 'p1' not in file and 'p2' not in file])
overpass_count = len([file for file in file_list if 'overpass' in file and 'p1' not in file and 'p2' not in file])
vertigo_count = len([file for file in file_list if 'vertigo' in file and 'p1' not in file and 'p2' not in file])
anubis_count = len([file for file in file_list if 'anubis' in file and 'p1' not in file and 'p2' not in file])
ancient_count = len([file for file in file_list if 'ancient' in file and 'p1' not in file and 'p2' not in file])
dust2_count = len([file for file in file_list if 'dust2' in file and 'p1' not in file and 'p2' not in file])

# Histogram of the map distribution
map_counts = [inferno_count, mirage_count, nuke_count, overpass_count, vertigo_count, anubis_count, ancient_count, dust2_count]
map_names = ['Inferno', 'Mirage', 'Nuke', 'Overpass', 'Vertigo', 'Anubis', 'Ancient', 'Dust2']

fig, ax = plt.subplots()
bars = ax.bar(map_names, map_counts)

ax.bar_label(bars, padding=1)

plt.xlabel('Map')
plt.ylabel('Count')
plt.title('Map distribution of the demos')
plt.show()
```
