### Create graph

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("./demo-parse/parse/player_frames/katowice2023_faze-vs-fnatic-m2-inferno.csv")
data2 = pd.read_csv("./demo-parse/parse/player_frames/katowice2023_g2-vs-big-m2-inferno.csv")
data3 = pd.read_csv("./demo-parse/parse/player_frames/katowice2023_g2-vs-faze-m1-inferno.csv")
data4 = pd.read_csv("./demo-parse/parse/player_frames/katowice2023_g2-vs-liquid-m2-inferno.csv")
data5 = pd.read_csv("./demo-parse/parse/player_frames/katowice2023_heroic-vs-natus-vincere-m1-inferno.csv")
data6 = pd.read_csv("./demo-parse/parse/player_frames/katowice2023_heroic-vs-og-m3-inferno.csv")


pts1 = data[['x','y','z']].copy()
pts2 = data2[['x','y','z']].copy()
pts3 = data3[['x','y','z']].copy()
pts4 = data4[['x','y','z']].copy()
pts5 = data5[['x','y','z']].copy()
pts6 = data6[['x','y','z']].copy()

pts = pd.concat([pts1, pts2, pts3, pts4, pts5, pts6])
pts['numpts'] = 0

pts
```

```python
from scipy.spatial.distance import cdist
import numpy as np

x = np.linspace(-2000,3000, 60)
y = np.linspace(-1000,4000, 60)

grid = np.column_stack((np.repeat(x, len(y)), np.tile(y, len(x))))

dens = []

# Density
for x_axis in range(-2000,3000,1000):
    for y_axis in range(-1000,4000,1000):
        temp_dens = pts.loc[(pts['x'] >= x_axis) & (pts['x'] < x_axis + 1000) & (pts['y'] >= y_axis) & (pts['y'] < y_axis + 1000)].shape[0]
        dens.append([x_axis, y_axis, temp_dens])

for d in dens:
    pts.loc[(pts['x'] >= d[0]) & (pts['x'] < d[0] + 1000) & (pts['y'] >= d[1]) & (pts['y'] < d[1] + 1000), 'numpts'] = d[2]

pts['dens'] = pts['numpts'] / len(pts)

pts
```

```python
bars = pts[['numpts', 'dens']].drop_duplicates()
bars['numpts'] = bars['numpts'].astype(str)
plt.bar(bars['numpts'], bars['dens'])
print(bars.sort_values(by=['dens']))
```

```python
from scipy.spatial.distance import cdist
import numpy as np
import math

x = np.linspace(-2000,3000, 60)
y = np.linspace(-1000,4000, 60)

grid = np.column_stack((np.repeat(x, len(y)), np.tile(y, len(x))))

# Function to calculate points inside circle
def count_points_in_circle(center, radius, points):
    distances = np.sqrt((points.iloc[:,0] - center[0]) ** 2 + (points.iloc[:,1] - center[1]) ** 2)
    points_inside_circle = np.sum(distances <= radius)
    return points_inside_circle

# Szűrés a grid pontokra
filtered_grid_points_ten = []
filtered_grid_points_fifteen = []
filtered_grid_points_twenty = []

for grid_point in grid:
    points_inside = count_points_in_circle(grid_point, 7, pts[['x', 'y']])
    if points_inside >= 10:
        filtered_grid_points_ten.append(grid_point)
    if points_inside >= 15:
        filtered_grid_points_fifteen.append(grid_point)
    if points_inside >= 20:
        filtered_grid_points_twenty.append(grid_point)

filtered_grid_points_ten = np.array(filtered_grid_points_ten)
filtered_grid_points_fifteen = np.array(filtered_grid_points_fifteen)
filtered_grid_points_twenty = np.array(filtered_grid_points_twenty)

graph = pd.DataFrame(filtered_grid_points_ten, columns = ['gX', 'gY'])
graph.to_csv('./graph-models/beta_0.6_1.csv', index=False)
graph = pd.DataFrame(filtered_grid_points_fifteen, columns = ['gX', 'gY'])
graph.to_csv('./graph-models/beta_0.6_2.csv', index=False)
graph = pd.DataFrame(filtered_grid_points_twenty, columns = ['gX', 'gY'])
graph.to_csv('./graph-models/beta_0.6_3.csv', index=False)
```

### Graph edges

```python
import pandas as pd
import numpy as np

graph = pd.read_csv('./graph-models/beta_v1.csv')
graph['idx'] = graph.index

edges = np.column_stack((np.repeat(graph['idx'].to_numpy(), len(graph['idx'].to_numpy())), np.tile(graph['idx'].to_numpy(), len(graph['idx'].to_numpy()))))

edge = pd.DataFrame(edges, columns = ['source', 'target'])
edge = edge.merge(graph.rename(columns = {'gX':'sourceX', 'gY': 'sourceY', 'idx': 'sourceIdx'}), left_on='source', right_on='sourceIdx')
edge = edge.merge(graph.rename(columns = {'gX':'targetX', 'gY': 'targetY', 'idx': 'targetIdx'}), left_on='target', right_on='targetIdx')
edge['neighbour'] = np.sqrt((edge['sourceX'] - edge['targetX']) ** 2 + (edge['sourceY'] - edge['targetY']) ** 2) <= 85
edge.drop(columns=['source', 'target'], inplace=True)
edge = edge.loc[edge['neighbour'] == True]

edge.to_csv('./graph-models/beta_v1_edges.csv', index=False)
```
